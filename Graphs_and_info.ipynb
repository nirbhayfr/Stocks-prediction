```python
import yfinance as yf
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px

def get_stock_prize_fig(df):
    fig = px.line(df, x="Date", y= ["Open","Close"], title="Closing and Opening Price vs Date")
    return fig

def get_more(df):
  df['EWA_20'] = df['Close'].ewm(span=20,adjust=False).mean()
  fig = px.scatter(df, x = "Date" , y = "EWA_20", title="Exponential Moving Average vs Date")
  fig.update_traces(mode="lines+markers" , marker=dict(color="red"), line=dict(color="blue"))
  return fig

val = "MSFT"  #STOCK CODE INPUT FROM HTML
stock_info = yf.Ticker(val) #RETURNS OBJECT OF INFORMATION

# print(stock_info.info) #RETURNS FULL DESCRIPTION

#To display data when a stock code is entered and submit is clicked
print(stock_info.info["website"]) #RETURNS COMPANY WEBSITE
print(stock_info.info["longBusinessSummary"]) #RETURNS PARAGRAPH (COMPANY SUMMARY)

#takes date from date input on the website and creates a dataframe
df = yf.download(val, '2024-08-01') #Start date will be taken from date input on the website
df.reset_index(inplace=True)
#print(df) #Prints the whole database

#creates a plot graph from the data frame when "stock prize" button is clicked
fig = get_stock_prize_fig(df)
fig.show()

#creates a scatter graph from the data frame when "indicators" button is clicked
fig2 = get_more(df)
fig2.show()
```
